```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()


# First two columns from iris.data
# CONVENTION x = features, y = target
X = iris.data[:, :2] 
# NOTE: (Last two colums is more accurate than the first two - see graphs)
# Howeverm in this example the first two columns are used.
y = iris.target

print iris.data
print X
print y

# Supervised learning - Supply the data and the results
# Classification problem - There is a target - discrete predictions 
# rather than regression with predicts a  more exact value
```

```python
# Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split

# The test size is 0.25 or 25% of the whole dataset. 
# A random state of one fixes the random seed of the function 
# so that you get the same results every time you call the 
# function, which is important for now to reproduce the same 
# results consistently.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
# Now load an estimator
from sklearn.svm import SVC

# Create an instance of a linear SVC (support vector classifier)
clf = SVC(kernel='linear',random_state=1)

# return a trained model
clf.fit(X_train, y_train)
# clf variable is now fitted/trained

# Generate predictions based on X
y_pred = clf.predict(X_test)
# Predicted y results
print(y_pred)
# Actual y results
print(y_test)

# Get accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
# SUMMARY 
#
# To summarize, we will do all the steps with a different algorithm, 
# logistic regression.

# Steps:
# 1. Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# 2. Split the data into training and testing sets
#    - All rows and FIRST two columns
X = iris.data[:, :2]
#    - Set y as the target
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

# 3. Create and train the model using the fit function
clf = LogisticRegression(random_state = 1)
clf.fit(X_train, y_train)

# 4. Predict the test observations
y_pred = clf.predict(X_test)

# 5. Measure the accuracy
accuracy_score(y_test,y_pred)
```
