```python
# 9 - Pandas
# A library built on top of numpy

# Enable matplotlib for Jupyter
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

# Load the dataset
iris = datasets.load_iris()

# Create a DATAFRAME with the observation variables
# iris.data and column names as arguments
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)

# Create a histogram of the values in the dataframe for 'sepal length'
iris_df['sepal length (cm)'].hist(bins=30)
```

```python
# OR: Create a historgram with a different colour for each feature name
for class_number in np.unique(iris.target):    
    plt.figure(1)    
    iris_df['sepal length (cm)'].iloc[np.where(iris.target == class_number)[0]].hist(bins=30)
```

```python
# And now with real data...

from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
target = iris.target

# Resize the figure for better viewing
plt.figure(figsize=(12,5))

# First subplot
plt.subplot(121)

# Visualize the first two columns of data:
# column 0 and column 1
plt.scatter(data[:,0], data[:,1], c=target)

# Second subplot
plt.subplot(122)

# Visualize the last two columns of data:
# The c parameter takes an array of colorsin this case, 
# the colors 0, 1, and 2 in the iris target:
# column 2 and 3 (x vs y)
plt.scatter(data[:,2], data[:,3], c=target)
```
