```python
# 10 - Matplotlib 
# A library built on top of numpy

# Enable matplotlib for Jupyter
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot(x variables, y varibales) 
# - x and y arrays must be the same size

plt.plot(np.arange(10)+1, np.arange(10))
```

```python
# Plot an exponential
plt.plot(np.arange(10), np.exp(np.arange(10)))
```

```python
# Plotting graphs side by side
plt.figure()
plt.subplot(121)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(122)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
# Or top to bottom:

plt.figure()
plt.subplot(211)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(212)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
# Note:
# 
# The first two numbers in the subplot command refer 
# to the grid size in the figure instantiated by plt.figure(). 
# The grid size referred to in plt.subplot(221) is 2 x 2, 
# the first two digits. The last digit refers to traversing 
# the grid in reading order: left to right and then up to down

plt.figure()
plt.subplot(221)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(222)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(223)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(224)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```
